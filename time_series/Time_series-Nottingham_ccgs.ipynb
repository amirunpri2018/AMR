```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns

#load datasets
ccgs_92377=pd.read_csv('ccgs_92377_clean.csv',low_memory=False)

#Nottingham
ccg1='Nottingham City'
ccg2='Nottingham North And East'
ccg3='Nottingham West' 
time_index=ccgs_92377[ccgs_92377['CCG']==ccg1]['Timeperiod']

ccgs_compare=pd.DataFrame()
ccgs_compare[ccg1]=ccgs_92377[ccgs_92377['CCG']==ccg1].loc[:,'Value'].values
ccgs_compare[ccg2]=ccgs_92377[ccgs_92377['CCG']==ccg2].loc[:,'Value'].values
ccgs_compare[ccg3]=ccgs_92377[ccgs_92377['CCG']==ccg3].loc[:,'Value'].values

datetime_index=pd.to_datetime(ccgs_92377[ccgs_92377['CCG']==ccg1]['Timeperiod'], format='%Y-%m-%d')

ccgs_compare=ccgs_compare.set_index(datetime_index)

print ('NHS Targets:\n 2013/14=1.161\n 2018/19=0.965')
sns.set(context='talk',style='ticks')
ax=ccgs_compare.plot(color=['blue','purple','orange'],figsize=(8,6))
ax.set_title('Nottingham Area CCGs vs. Targets', fontsize=18)
ax.set_xlabel('Month', fontsize=16)
ax.set_ylabel('Values', fontsize=16)
ax.axhline(y=1.161, color='green', linestyle='--', label='2013/14 Target') #Targets: 2013/14= 1.161, 2018/19=0.965
ax.axhline(y=0.965, color='red', linestyle='--', label='2018/19 Target') 
ax.axvline(x='Nov 2015', color='yellow', linestyle='dotted') 
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
